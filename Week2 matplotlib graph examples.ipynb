```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
#DATA
x=np.arange(0,10,0.1)
```

```python
y=np.sin(x)
```

```python
plt.plot(x,y)
plt.title("sinüs grafiği")
plt.xlabel("X Ekseni")
plt.ylabel("Y Ekseni")
plt.show()
```
